## 1. Initial Imports + Data Collection

### We import the pandas and NumPy libraries for data collection, cleaning, and analysis of the dataset and json

```python
import numpy as np
import pandas as pd
import json
```

```python
df = pd.read_csv('games.csv')
```

#### Observe the types of columns that we have... 
#### For the most part, if attempting to gauge which team shall win off of events that happen earlier on in the game, FOR NOW we would be interested in the champions both teams have, as well as who the winner ended up being

```python
df.columns
```

```python
df['seasonId'].value_counts() # all data is from season 9 so not relevant to include this column in the DataFrame
```

```python
champion_data = json.load(open('champion_info.json'))
```

```python
champion_data # data format below
```

```python
df['team1_champ1'] = df['t1_champ1id'].apply(lambda id: champion_data['data'][str(id)]['name'])
df['team1_champ2'] = df['t1_champ2id'].apply(lambda id: champion_data['data'][str(id)]['name'])
df['team1_champ3'] = df['t1_champ3id'].apply(lambda id: champion_data['data'][str(id)]['name'])
df['team1_champ4'] = df['t1_champ4id'].apply(lambda id: champion_data['data'][str(id)]['name'])
df['team1_champ5'] = df['t1_champ5id'].apply(lambda id: champion_data['data'][str(id)]['name'])

df['team2_champ1'] = df['t2_champ1id'].apply(lambda id: champion_data['data'][str(id)]['name'])
df['team2_champ2'] = df['t2_champ2id'].apply(lambda id: champion_data['data'][str(id)]['name'])
df['team2_champ3'] = df['t2_champ3id'].apply(lambda id: champion_data['data'][str(id)]['name'])
df['team2_champ4'] = df['t2_champ4id'].apply(lambda id: champion_data['data'][str(id)]['name'])
df['team2_champ5'] = df['t2_champ5id'].apply(lambda id: champion_data['data'][str(id)]['name'])
```

```python
df = df[['team1_champ1', 'team1_champ2', 'team1_champ3', 'team1_champ4', 'team1_champ5', 'team2_champ1', 'team2_champ2', 'team2_champ3', 'team2_champ4', 'team2_champ5','firstBlood','firstTower','firstDragon','firstBaron','winner']]
```

```python
df
```

```python
t1_champ_binary = [pd.get_dummies(df[col], prefix='t1') for col in ['team1_champ1','team1_champ2','team1_champ3','team1_champ4','team1_champ5']]
t1_binary_df = sum(t1_champ_binary)

t2_champ_binary = [pd.get_dummies(df[col], prefix='t2') for col in ['team2_champ1','team2_champ2','team2_champ3','team2_champ4','team2_champ5']]
t2_binary_df = sum(t2_champ_binary)
```

```python
df = df.join(t1_binary_df).join(t2_binary_df)
```

```python
df
```

```python
df = df.drop(['team1_champ1', 'team1_champ2', 'team1_champ3', 'team1_champ4', 'team1_champ5', 'team2_champ1', 'team2_champ2', 'team2_champ3', 'team2_champ4', 'team2_champ5'], axis=1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
X, y = df.drop('winner', axis=1), df['winner']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf = RandomForestClassifier(n_jobs = -1)
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
